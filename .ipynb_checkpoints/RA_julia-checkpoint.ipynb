```julia
using CSV
using DataFrames
using GLM
using Plots
```

```julia
df = CSV.read("wage2015_subsample_inference.csv", DataFrame);
first(df, 10)
```

```julia
df_scl = filter(:scl => ==(1), df)
```

```julia
df_clg = filter(:clg => ==(1), df)
```

```julia
sample = [df_scl;df_clg]
```

```julia
linear_m = lm(@formula(lwage ~ sex + exp1 + shs + hsg + mw + so + we + occ2 + ind2), sample)
```

```julia
linear_m_flex = lm(@formula(lwage ~ sex + exp1 + exp1*exp2 + exp1*exp3 + exp1*exp4 + exp1*hsg + exp1*ind2 + exp1*mw + exp1*occ2 + exp1*shs + exp1*so + exp1*we +
        exp2 + exp2*exp3 + exp2*exp4 + exp2*hsg + exp2*ind2 + exp2*mw + exp2*occ2 + exp2*shs + exp2*so + exp2*we +
        exp3 + exp3*exp4 + exp3*hsg + exp3*ind2 + exp3*mw + exp3*occ2 + exp3*shs + exp3*so + exp3*we +
        exp4 + exp4*hsg + exp4*ind2 + exp4*mw + exp4*occ2 + exp4*shs + exp4*so + exp4*we +
        hsg + hsg*ind2 + hsg*mw + hsg*occ2 + hsg*shs + hsg*so + hsg*we +
        ind2 + ind2*mw + ind2*occ2 + ind2*shs + ind2*so + ind2*we +
        mw + mw*occ2 + mw*shs + mw*so + mw*we +
        occ2 + occ2*shs + occ2*so + occ2*we +
        shs + shs*so + shs*we +
        so + so*we +
        we), sample)
```

```julia
function coefplot(lm, lmf)
    n = ["Basic Model", "      Flexible Model"] # no intercept
    vals = [coef(lm)[2], coef(lmf)[2]]
    # errors = stderr(m)[2:end]
    errors = [stderror(lm)[2], stderror(lmf)[2]]
    scatter(n, vals,
        legend = false,
        yerror = 1.96 .* errors,
        title = "Interval confidence of sex coefficient",
        xrotation = 90,
        xlim = (0.3, 1.8),
        ylim = (-0.13, 0)
    )
end
```

```julia
using PyCall
np = pyimport("numpy")
```

```julia
fig, ax = plt.subplots(figsize=(10, 7))

ax = plt.gca()
ax.set_ylim([2, 3.5])
ax.set_xlim([0, 40])

plt.scatter(df_scl[!, "exp1"], df_scl[!, "lwage"], color=(0.1, 0.2, 0.8, 0.05), label="_nolegend_")

z = np.polyfit(df_scl[!, "exp1"], df_scl[!, "lwage"], 4)
p = np.poly1d(z)

linea_re= DataFrame("x" => df_scl[!, "exp1"],
                    "y"=> p(df_scl[!, "exp1"]))

linea_re = sort(linea_re, [order("x")])

plt.plot(linea_re[!, "x"], linea_re[!, "y"], "--", color=(0.1, 0.2, 0.8))

ax.annotate("College Graduated",
    xy = (10, 3.15),
    xytext = (5, 3.27),
    arrowprops = Dict("facecolor" => "black",
        "shrink" => 0.01))




plt.scatter(df_clg[!, "exp1"], df_clg[!, "lwage"], color=(0.8, 0.2, 0.1, 0.05), label="_nolegend_")

z = np.polyfit(df_clg[!, "exp1"], df_clg[!, "lwage"], 4)
p = np.poly1d(z)


linea_re= DataFrame("x" => df_clg[!, "exp1"],
    "y" => p(df_clg[!, "exp1"]))

linea_re = sort(linea_re, [order("x")])

plt.plot(linea_re[!, "x"], linea_re[!, "y"], "--", color=(0.8, 0.2, 0.1))


ax.annotate("Some College",
    xy = (9, 2.779),
    xytext = (7, 2.6),
    arrowprops = Dict("facecolor" => "black",
        "shrink" => 0.01))


plt.ylabel("Log Wage")
plt.xlabel("Years of experience")

plt.show()
```
